```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat = pd.read_csv('datos_procesados.csv')
```

```python
plt.hist(dat['age'], bins=15, color='skyblue', edgecolor='black')

plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades')

plt.show()
```

```python
hom_data = dat[dat['sex'] == 1]  # 1 representa a hombres
muj_data = dat[dat['sex'] == 0]  # 0 representa a mujeres
# Calcular la cantidad de anémicos, diabéticos, fumadores y muertos por género
male_counts = [
    hom_data['anaemia'].sum(),
    hom_data['diabetes'].sum(),
    hom_data['smoking'].sum(),
    hom_data['DEATH_EVENT'].sum()
]
female_counts = [
    muj_data['anaemia'].sum(),
    muj_data['diabetes'].sum(),
    muj_data['smoking'].sum(),
    muj_data['DEATH_EVENT'].sum()
]
color_hom = 'blue'
color_muj = 'red'

categories = ['Anémicos', 'Diabéticos', 'Fumadores', 'Muertos']
bar_width = 0.35
index = range(len(categories))
plt.bar(index, male_counts, bar_width, label='Hombres', color=color_hom)
plt.bar([i + bar_width for i in index], female_counts, bar_width, label='Mujeres', color=color_muj)
plt.xlabel('Categorías')
plt.ylabel('Cantidad')
plt.title('Comparación por género')
plt.xticks([i + bar_width / 2 for i in index], categories)
plt.legend()
plt.tight_layout()
plt.show()
```
