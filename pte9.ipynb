```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import plotly.graph_objects as go
import plotly.express as px
```

```python
objetivo = df['DEATH_EVENT']
arr_objetivo = objetivo.values
```

```python
y = arr_objetivo
```

```python
X_embedded = TSNE(
    n_components=3,
    learning_rate='auto',
    init='random',
    perplexity=3
).fit_transform(y.reshape(-1, 1))

dat = {'Dim_1': X_embedded[:, 0], 'Dim_2': X_embedded[:, 1], 'Dim_3': X_embedded[:, 2], 'DEATH_EVENT': y}
dt = pd.DataFrame(dat)
col_map = {0: 'Muerto', 1: 'Vivo'}
dt['Color'] = dt['DEATH_EVENT'].map(col_map)

fig = px.scatter_3d(dt, x='Dim_1', y='Dim_2', z='Dim_3', color='Color',
                    title='Grafico Dispersion 3D con Color por Clase')

fig.update_traces(marker=dict(size=4, opacity=0.8))
fig.show()
```
