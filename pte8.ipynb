```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('graficos/datos_procesados.csv')
```

```python

# Supongamos que df es tu DataFrame con las columnas adecuadas

# Filtrar por categoras
anaemic = df[df['anaemia'] == 1]
non_anaemic = df[df['anaemia'] == 0]

diabetic = df[df['diabetes'] == 1]
non_diabetic = df[df['diabetes'] == 0]

smoker = df[df['smoking'] == 1]
non_smoker = df[df['smoking'] == 0]

dead = df[df['DEATH_EVENT'] == 1]
alive = df[df['DEATH_EVENT'] == 0]

# Configurar subgrficas
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Grfica para Anaemia
axes[0, 0].pie([len(anaemic), len(non_anaemic)], labels=['S', 'No'], autopct='%1.1f%%', startangle=90)
axes[0, 0].set_title('(Anmicos)')

# Grfica para Diabetes
axes[0, 1].pie([len(diabetic), len(non_diabetic)], labels=['S', 'No'], autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title('(Diabticos)')

# Grfica para Fumadores
axes[1, 0].pie([len(smoker), len(non_smoker)], labels=['S', 'No'], autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('(Fumadores)')

# Grfica para Muertos
axes[1, 1].pie([len(dead), len(alive)], labels=['S', 'No'], autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('(Muertos)')

plt.tight_layout()
plt.show()
```
